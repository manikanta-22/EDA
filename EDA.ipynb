```python
cols_dict=dict(visa_dataset.dtypes)

#=============================M-1====================================
cat_list,num_list=[],[]
for key,value in cols_dict.items():
    if value=='O':
        cat_list.append(key)
    else:
        num_list.append(key)
cat_list,num_list


#===============================M-2====================================
cat_list=[key for key,value in cols_dict.items() if value=='O']

num_list=[key for key,value in cols_dict.items() if value!='O']


#pandas
#dictionary
#for
#if
#list

```
